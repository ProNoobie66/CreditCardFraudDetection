```python
#credit card fraud detection
#Importing libraries 
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('credit card fraud data.csv') #importing and saving data in dataframe
```

```python
credit_card_data.shape #showing number of columns and rows
```

```python
credit_card_data.tail() #showing 5 rows
```

```python
credit_card_data.info() # telling which type data is stored like float,int etc.
```

```python
credit_card_data.isnull().sum() #checking how much data is missing 
```

```python
credit_card_data['Class'].value_counts() #it gives how many fraud and legit transaction are there
```

```python
legit = credit_card_data[credit_card_data.Class == 0] # dividing legit and fraud transaction data in different dataframe
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape) #checking how many columns and rows are there in legit and fraud
print(fraud.shape)
```

```python
legit.Amount.describe() #describing amount data of legit
```

```python
fraud.Amount.describe() ##describing amount data of fraud
```

```python
credit_card_data.groupby('Class').mean() #mean of legit and fraud in different group
```

```python
legit_sample = legit.sample(n=600) #taking 600 data from legit to make data size of fraud and legit transaction comparable
# In logistic regression if data size of one output is very large than other then problem may happen by under fitting data 
```

```python
legit_sample.shape # seeing number of row and column in new legit sample
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0) #combining fraud and legit sample
```

```python
new_dataset.head() #New dataset 5 rows
```

```python
new_dataset.shape # seeing number of row and column in new data set
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts() #It give number of legit and fraud transaction
```

```python
new_dataset.groupby('Class').mean() 
```

```python
X = new_dataset.drop(columns='Class', axis=1) #droping column class so that we have all inputs in x
Y = new_dataset['Class'] #column class is our output hence putting column class in y
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) #spliting data in training and test data set
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression() #using  logistic regression and saving in model
```

```python
model.fit(X_train, Y_train) #training model using train data set
```

```python
X_train_prediction = model.predict(X_train) #checking accuracy for the training data set
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test) #checking accuracy for the test data set
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
